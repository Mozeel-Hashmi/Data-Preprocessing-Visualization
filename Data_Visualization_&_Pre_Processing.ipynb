```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def f(x, y):
    return x**2 + y**2

def grad_f(x, y):
    df_dx = 2 * x
    df_dy = 2 * y
    return np.array([df_dx, df_dy])

x0, y0 = 4.0, 4.0
alpha = 0.1
iterations = 50

path = [(x0, y0, f(x0, y0))]

x, y = x0, y0
for _ in range(iterations):
    grad = grad_f(x, y)
    x -= alpha * grad[0]
    y -= alpha * grad[1]
    path.append((x, y, f(x, y)))

path = np.array(path)

x_vals = np.linspace(-5, 5, 100)
y_vals = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X, Y)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.7, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
ax.plot(path[:, 0], path[:, 1], path[:, 2], color='red', marker='o')
ax.set_title('3D Gradient Descent')
plt.show()
```

### Removing Outliers
Outliers can be removed using various techniques, including:
1. Statistical Methods
2. Model-Based Approaches
3. Visual Inspection
4. Domain Knowledge
5. Winsorization

For this task, Ill focus on just two statistical methods for outlier removal. One is z-score and the second is IQR method.

#### z-score
Using the Z-score method, we can both detect and remove outliers from our data.
A data point is typically considered an outlier if its Z-score is greater than 3 or less than -3, meaning it lies far from the mean in a standard normal distribution.

```python
# Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Loading data
a = sns.load_dataset('diamonds')
df = pd.DataFrame(a)

# Detecting outliers using z-score and removing them
z_score = np.abs(zscore(df['carat']))
outliers = df[z_score > 3]
clean_data = df[z_score < 2.5]

# Boxplot to compare the cleaned and uncleaned data

# Uncleaned Data
plt.figure(figsize = (18, 6))
plt.subplot(1, 2, 1)
sns.boxplot(y = df['carat'])
plt.title('Before')

# Cleaned Data
plt.subplot(1, 2, 2)
sns.boxplot(y = clean_data['carat'])
plt.title('After')
plt.suptitle('Uncleaned vs Cleaned (Z-Score)')
plt.show()
```

#### IQR-Method
Outliers can also be removed using the Interquartile Range (IQR) method, which is particularly effective when the data is skewed or not normally distributed.

```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loading data
a = sns.load_dataset('diamonds')
df = pd.DataFrame(a)

# Detecting and Removing outliers using IQR-Method
Q1 = df['carat'].quantile(0.25)
Q3 = df['carat'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR
outliers = df[(df['carat'] < lower_bound) | (df['carat'] > upper_bound)]
cleaned_data = df[(df['carat'] > lower_bound) & (df['carat'] < upper_bound)]

# Visully representing the comparision
plt.figure(figsize = (18, 6))
c = '#8227f2'
plt.subplot(1, 3, 1)
sns.boxplot(y = df['carat'], color = c)
plt.title('Before')
plt.subplot(1, 3, 2)
sns.boxplot(y = cleaned_data['carat'], color = c)
plt.title('After')
plt.suptitle('Uncleaned vs Cleaned (IQR-Method)')
plt.subplot(1, 3, 3)
sns.scatterplot(x = cleaned_data['carat'], y = np.random.normal(7, 0.2, size = cleaned_data.shape[0]), label = 'Clean', alpha = 0.7, color = c)
sns.scatterplot(x = outliers['carat'], y = np.random.normal(7, 0.2, size = outliers.shape[0]), label = 'Outliers', c = 'r', alpha = 0.5)
plt.title('Scatter Plot')
plt.show()
```

#### Cap
In capping we donot delete the outliers we just replace their values with the most suitable maximum or minimum value. By doing so the shape of our data remain intact and the effect of outliers is reduced.

```python
# Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Loading Data
a = sns.load_dataset('diamonds')
df = pd.DataFrame(a)

# IQR to find Upper and Lower bound
Q1 = df['carat'].quantile(0.25)
Q3 = df['carat'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Capping outliers
df ['carat_capped'] = df['carat'].clip(lower = lower_bound, upper = upper_bound)

# Plotting Boxplot
plt.figure(figsize = (18, 6))
c = '#0addf0'
plt.subplot(1, 2, 1)
sns.boxplot(y = df['carat'], color = c)
plt.title('Before Capping')
plt.subplot(1, 2, 2)
sns.boxplot(y = df['carat_capped'], color = c)
plt.title('After Capping')
plt.suptitle('Capping')
plt.show()
```

## Data Binning
Data binning (also known as discretization) is a data preprocessing technique used to group continuous data into intervals or bins. Instead of working with raw, continuous values, you divide the data range into a finite number of bins and assign each value to one of these bins. There are different types of binning. 4 of them are as follow:
1. Equal Width Binning
2. Equal Frequency Binning (Quantile Binning)
3. Custom Binning
4. Clustring-Based Binning

### Equal Width Binning
It divides the entire range of a continuous variable into intervals of equal width, meaning each bin spans the same range of values, regardless of how many data points fall into each bin. This technique helps simplify the distribution by grouping similar values together based on their position within the range. pd.cut() function is used to do Equal Width Binning. It is used when data is normally or uniformly distributed

```python
# Importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Loading data
a = sns.load_dataset('diamonds')
df = pd.DataFrame(a)

# binning 
x = df['carat'].values
df['carat_binned'] = pd.cut(x, bins = 20)
bins = df['carat_binned'].cat.categories
print(df.carat_binned.unique()) # To print overall bins created

# Visually showing how many data points are contained in each bin
palette = sns.color_palette("hsv", len(bins))
sns.countplot(x = df['carat_binned'], palette = palette)
plt.xticks(rotation = 45)
plt.title('Equal Width Binning (pd.cut())')
plt.show()
```

### Equal Frequency Binning
This method divides the data so that each bin has approximately the same number of observations. We use it when we want each bin to contain the same number of data points (e.g., quartiles, deciles). It is used when data is skewed

```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading data
a = sns.load_dataset('diamonds')
df = pd.DataFrame(a)

# binning (Equal Frequency using qcut)
df['carat_binned'] = pd.qcut(df['carat'], q = 20)
bins = df['carat_binned'].cat.categories

# Visullay showing bins
palette = sns.color_palette('hsv', len(bins))
sns.countplot(x = df['carat_binned'], palette = palette)
plt.xticks(rotation = 45)
plt.title('Equal Frequency Binning (pd.qcut())')
plt.show()
```

### Custom Binning
We use custom binning when we have a deep understanding of the data. This knowledge allows us to divide the data into meaningful intervals that make sense for the analysis and add real value.

```python
# importing libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# loading data
a = sns.load_dataset('diamonds')
df = pd.DataFrame(a)

# Custom binning (according to the data set the minimum value of carat is 0.200 and the max is 5.01)
bins = [0.2, 2, 4, 5.01]
label = ['Small', 'Medium', 'Big']
df['carat_binned'] = pd.cut(df['carat'], bins = bins, labels = label) # binning
bins = df['carat_binned'].cat.categories

# Visually representing bins
palette = sns.color_palette('hsv', len(bins))
sns.countplot(x = df['carat_binned'], palette = palette)
plt.title('Custom Binning')
plt.show()
```

### Cluster Binning
In cluster-based binning, we use machine learning algorithms to group values into clusters based on similarity, and then assign bins accordingly. It's more advanced and often more effective than traditional binning methods. Clustering is typically performed using libraries like scikit-learn.

```python
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# loading data
a = sns.load_dataset('diamonds')
df = pd.DataFrame(a)

# binning
kmeans = KMeans(n_clusters = 20, random_state = 0)
df['carat_binned'] = kmeans.fit_predict(df[['carat']])

# Visually showing bins
palette = sns.color_palette('hsv', n_colors = df['carat_binned'].nunique())
sns.countplot(x = df['carat_binned'], palette = palette)
plt.title('Cluster Binning')
plt.show()
```

## Data Scaling
Scaling is a technique used to normalize the range of independent variables (features) in your dataset. Two main types of scaling are as follow:
1. MinMax Scaler
2. Standard Scaler

### MinMax Scaler
MinMax Scaler scales the data to a fixed range, usually [0, 1].

```python
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Loading data
a = sns.load_dataset('diamonds')
df = pd.DataFrame(a)

# Initializing scaler and scaling data (atleast one feature)
scaler = MinMaxScaler()
df['carat_scaled'] = scaler.fit_transform(df[['carat']])
print(df['carat_scaled'])
```

### Standard-Scaler
Centers the data around the mean = 0, with standard deviation = 1. This scaling technique effects the mean. It makes the mean to be equal to 0.

```python
# Importing libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Loading Data
a = sns.load_dataset('diamonds')
df = pd.DataFrame(a)

# Initializing scaler and scaling data
scaler = StandardScaler()
df['carat_scaled'] = scaler.fit_transform(df[['carat']])
# print(df['carat_scaled'])

# Visually showing distribution
plt.figure(figsize = (18, 6))
plt.subplot(1, 2, 1)
sns.histplot(x = 'carat', data = df, color = 'skyblue', ec = 'black', kde = True)
plt.axvline(x = np.mean(df['carat']), ls = ':', color = 'red', label = 'Mean')
plt.axvline(x = np.median(df['carat']), ls = '--', color = 'green', label = 'Median')
plt.title('Before')
plt.legend()
plt.subplot(1, 2, 2)
sns.histplot(x = 'carat_scaled', data = df, color = 'skyblue', ec = 'black', kde = True)
plt.axvline(x = np.mean(df['carat_scaled']), ls = ':', color = 'red', label = 'Mean')
plt.axvline(x = np.median(df['carat_scaled']), ls = '--', color = 'green', label = 'Median')
plt.title('After')
plt.suptitle('Effect of Standard-Scaler on Mean')
plt.legend()
plt.show()
```

## Balancing Data (smote)
SMOTE stands for Synthetic Minority Oversampling Technique. It is used to balance the dataset by generating synthetic (new, fake but realistic) samples of the minority class.

```python
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# Loading data
df = pd.read_csv('Data/electricity_cost_dataset.csv')

# # Creating x and y
x = df.drop('structure type', axis = 1)
y = df['structure type']

# Splitting data into train test
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.2, stratify = y)

# balancing data using SMOTE
smote = SMOTE(random_state = 42)
X_resampled, Y_resampled = smote.fit_resample(X_train, Y_train)

# Showing the difference in values
print("Before:\n", y.value_counts())
print("After:\n", Y_resampled.value_counts())

# Creating Palette
clas = y.nunique()
palette = sns.color_palette('cool', clas)

# Comparision before and after smote
plt.figure(figsize = (12, 5))
plt.subplot(1, 2, 1)
sns.countplot(x = df['structure type'], palette = palette)
plt.title('Before')
plt.subplot(1, 2, 2)
sns.countplot(x = Y_resampled, palette = palette)
plt.title('After')
plt.suptitle('Comparision of with and without SMOTE')
plt.show()
```

## Train-Test-Split
train_test_split is a function from sklearn.model_selection used to split your dataset into training and testing sets. This is essential for machine learning because you want to train your model on one part of the data and test how well it performs on unseen data.

```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Loading Data
df = pd.read_csv('Data/electricity_cost_dataset.csv')

# Creating x and y
x = df.drop('structure type', axis = 1)
y = df['structure type']

# Spliting data into training and testing
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.2, stratify = y)

# Visualization of Training and Testing data

# Class Distribution
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.countplot(x=Y_train, color = 'skyblue', ec = 'black')
plt.title('Class Distribution in Training Data')
plt.subplot(1, 2, 2)
sns.countplot(x=Y_test, color = 'skyblue', ec = 'black')
plt.title('Class Distribution in Testing Data')
plt.tight_layout()
plt.show()
```

### Visualizing

```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Loading Data
df = pd.read_csv('Data/electricity_cost_dataset.csv')

# Creating x and y
x = df.drop('structure type', axis = 1)
y = df['structure type']

# Spliting data into training and testing
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.2, stratify = y)

# Distribution of data in training and testing
num_col = x.select_dtypes(include = 'number').columns
plt.figure(figsize = (20, 10))

for i, col in enumerate(num_col, 1):
    plt.subplot(2, 4, i)
    sns.kdeplot(X_train[col], label = 'Train', fill = True)
    sns.kdeplot(X_test[col], label = 'Test', fill = True)
    plt.title(f'Distribution of {col}')
    plt.legend()

plt.suptitle('Train vs Test KDE Distribution', fontsize = 18)
plt.tight_layout()
plt.show()
```
